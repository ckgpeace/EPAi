#### Checking the greater than function
`__gt__` function represented in polygon_one module `>`


```
# Example1
p1 = pol1.polygon1(n_edge = 7, r_circum = 10)
p2 = pol1.polygon1(n_edge = 4, r_circum = 9)

p1 > p2
```

```
# Example2
p1 = pol1.polygon1(n_edge = 4, r_circum = 10)
p2 = pol1.polygon1(n_edge = 4, r_circum = 9)

p1 > p2
```

```
# Example3 : less than equality also works - "Miracle - Miracle"
p1 = pol1.polygon1(n_edge = 4, r_circum = 10)
p2 = pol1.polygon1(n_edge = 10, r_circum = 1)

p1 < p2
```

```
# Example4
p1 = pol1.polygon1(n_edge = 4, r_circum = 10)
p2 = pol1.polygon1(n_edge = 10, r_circum = 1)

p1 > p2
```

```
import polygon2 as pol2
```

```
# Example 0
p1 = pol2.polygon2(6)

print(p1)

print(list(p1))
```

```
# Example 1
# The maximum efficiency for polygon of max 25 vertices
n = 26
p1 = pol2.polygon2(n)
print(f'The polygon with highest efficiency where max vertices are {len(list(p1))-1} i.e. max ratio of <Area/Perimeter> is {list(p1).index(max(list(p1)))}')
```

```
# Example 2
n = 6
p1 = pol2.polygon2(n)
print(f'The polygon with highest efficiency where max vertices are {len(list(p1))-1} i.e. max ratio of <Area/Perimeter> is {list(p1).index(max(list(p1)))} \n')
print("Below is the list of polygons with sides and area:perimeter ratio")
x = [print(i,r) for i,r in enumerate(list(p1)) if i>2]
```
